### Matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
filename = '/datasets/devset/dev-set/frames/video10-frame-1.jpg'
```

```python
image = mpimg.imread(filename)
```

```python
plt.axis('off')
plt.imshow(image)
plt.show()
```

### OpenCV

```python
import cv2
```

```python
image = cv2.imread(filename)
```

```python
plt.axis('off')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
